```python
a = sns.countplot(x = 'Class', data = data, palette = 'winter')
a.set_title('Countplot for 2 classes')
```

```python
data['Class'].value_counts()
```

### Boxplots for all features

```python
fig, axes = plt.subplots(nrows = 6, ncols = 2, figsize = (20, 40))
for nr, c in enumerate(data.columns):
    if c != 'Class':
        sns.boxplot(data = data, x = 'Class', y = c, ax = axes[nr // 2, nr % 2], palette = 'winter')
        axes[nr //2, nr % 2].set_title('Boxplot for ' + c + ' in pumpkins classes')
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
```
