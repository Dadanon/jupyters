### Start counplotting for each feature

```python
def plot(i, c):
    minm = X[c].min() # Find min
    maxm = X[c].max() # Find max
    q = (maxm - minm) / 5 # Create delta chunk
    arr = [minm + n * q for n in range(6)] # Create array of sections separated by q
    # Take each X[c], divide in to arr intervals and build a graph for each chunk
    a = sns.countplot(x = pd.cut(X[c], arr), data = X, palette = 'winter', ax = axes[i // 2, i % 2]) 
    a.set_title('Countplot for ' + c)
```

```python
fig, axes = plt.subplots(nrows = 6, ncols = 2, figsize = (20, 47))
for i in range(12):
    b = cols[i] # Take a separate column
    plot(i, b) # Equals to: plot(X[cols[i]], cols[i]) - plots len(arr) graphs with distribution for each separate feature
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
```

### Try to train our model after dropping Convex_Area and Solidity

Why we don't drop Equiv_Diameter? Don't know :)

```python
X1 = X.drop(columns = ['Solidity', 'Convex_Area'], axis = 1)
y1 = data['Class']
X1.head()
```

### Model selection

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size = 0.2)
```

### Here we (as I assume) will compare performance of same models with different datasets
#### (one with 'redundant' features and one without)

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
def model_evaluation(classifier):
    cm = confusion_matrix(y_test, classifier.predict(X_test)) # Make confusion matrix with true and pred values
    counts = [value for value in cm.flatten()] # Make array of counts from flattened cm
    labels = [f'{v1}' for v1 in counts] # Counts to str
    labels = np.asarray(labels).reshape(2, 2) # Reshape labels
    sns.heatmap(cm, annot = labels, cmap = 'Greens', fmt = '') # Maps values with colors and values
    y_pred = classifier.predict(X_test) # Predicted y on testset (after training)
    yt_pred = classifier.predict(X_train) # Predicted y on trainset (during training)
    print('The training accuracy of the algorithm is: ', accuracy_score(y_train, yt_pred)) # Displays training accuracy
    print('The testing accuracy of the algorithm is: ', accuracy_score(y_test, y_pred)) # Displays tested accuracy
    # Returns 2 accuracies, trained and tested in percents (why do we multiply by 100 twice?)
    return [(accuracy_score(y_train * 100, yt_pred * 100) * 100), (accuracy_score(y_test * 100, y_pred * 100) * 100)]
```

```python
def model_evaluation1(classifier):
    cm = confusion_matrix(y1_test, classifier.predict(X1_test))
    counts = [value for value in cm.flatten()]
    labels = [f'{v1}' for v1 in counts]
    labels = np.asarray(labels).reshape(2, 2)
    sns.heatmap(cm, annot = labels, cmap = 'Greens', fmt = '')
    y1_pred = classifier.predict(X1_test)
    yt1_pred = classifier.predict(X1_train)
    print('The training accuracy of the algorithm is: ', accuracy_score(y1_train, yt1_pred))
    print('The testing accuracy of the algorithm is: ', accuracy_score(y1_test, y1_pred))
    return [(accuracy_score(y1_train * 100, yt1_pred * 100) * 100), (accuracy_score(y1_test * 100, y1_pred * 100) * 100)]
```

### K-nearest neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(10, weights = 'uniform', metric = 'manhattan')
knn.fit(X_train, y_train)
model_evaluation(knn)
```

```python
knn1 = KNeighborsClassifier(10, weights = 'uniform', metric = 'manhattan')
knn1.fit(X1_train, y1_train)
model_evaluation1(knn1)
```

### Support Vector Classifier

```python
from sklearn.svm import SVC
```

```python
svc = SVC(C = 1E6)
svc.fit(X_train, y_train)
model_evaluation(svc)
```

```python
svc1 = SVC(C = 1E6)
svc1.fit(X1_train, y1_train)
model_evaluation1(svc1)
```

### Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(max_depth=6)
dt.fit(X_train, y_train)
model_evaluation(dt)
```

```python
dt1 = DecisionTreeClassifier(max_depth=6)
dt1.fit(X1_train, y1_train)
model_evaluation1(dt1)
```

### Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(max_depth = 4)
rf.fit(X_train, y_train)
model_evaluation(rf)
```

```python
rf1 = RandomForestClassifier(max_depth = 4)
rf1.fit(X1_train, y1_train)
model_evaluation1(rf1)
```

### Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier(learning_rate = 0.005)
gbc.fit(X_train, y_train)
model_evaluation(gbc)
```

```python
gbc1 = GradientBoostingClassifier(learning_rate = 0.005)
gbc1.fit(X1_train, y1_train)
model_evaluation1(gbc1)
```

```python
results1 = pd.DataFrame(
    columns = [
        'Training Accuracy Base Model',
        'Testing Accuracy Base Model',
    ],
    index = [
        'K-Nearest Neighbors',
        'Support Vector Machines',
        'Decision Tree',
        'Random Forest',
        'Gradient Boost',
    ]
)
```

```python
results1.loc['K-Nearest Neighbors'] = model_evaluation(knn);
results1.loc['Support Vector Machines'] = model_evaluation(svc);
results1.loc['Decision Tree'] = model_evaluation(dt);
results1.loc['Random Forest'] = model_evaluation(rf);
results1.loc['Gradient Boost'] = model_evaluation(gbc);
```

```python
results1
```

```python
results2 = pd.DataFrame(
    columns = [
        'Training Accuracy Dropped Columns',
        'Testing Accuracy Dropped Columns',
    ],
    index = [
        'K-Nearest Neighbors',
        'Support Vector Machines',
        'Decision Tree',
        'Random Forest',
        'Gradient Boost',
    ]
)
```

```python
results2.loc['K-Nearest Neighbors'] = model_evaluation1(knn1);
results2.loc['Support Vector Machines'] = model_evaluation1(svc1);
results2.loc['Decision Tree'] = model_evaluation1(dt1);
results2.loc['Random Forest'] = model_evaluation1(rf1);
results2.loc['Gradient Boost'] = model_evaluation1(gbc1);
```

```python
results2
```

```python
results = pd.concat([results1, results2], join = 'outer', axis = 1)
```
